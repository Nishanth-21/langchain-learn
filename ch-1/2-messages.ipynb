```python
# from google.genai._gaos.utils import retries
import os

from langchain.chat_models import init_chat_model
from langchain.messages import AIMessage,HumanMessage,SystemMessage

os.environ["GROQ_API_key"] = os.getenv("GROQ_API_key")

llm_groq = init_chat_model(

    model="qwen/qwen3-32b",
    model_provider="groq",
    temperature = 0.5,
    
    max_retries = 5,
    timeout = 120

)

myMessages =[
    
    SystemMessage(content="you are a genz, who answers in funny way"),
    HumanMessage(content="tell me a fun fact about cats")
]


response=llm_groq.invoke(myMessages)
print(response.content_blocks)
# print(llm_groq)
```

```python
from langchain_core.prompts import PromptTemplate

user_input = input("enter the topic:")

dynamic_prompt = PromptTemplate.from_template("write a fun fact about {topic}")

ready_prompt = dynamic_prompt.invoke({"topic":user_input})
response = llm_groq.invoke(ready_prompt)
response.content
```

```python
user_input = input("enter the query:")
dynamic_prompt = PromptTemplate.from_template("{topic}")
ready_prompt = dynamic_prompt.invoke({"topic":user_input})

response = llm_groq.invoke(ready_prompt)
print(response.content)
```

```python
from langchain_core.prompts import ChatPromptTemplate

user_input = input("enter the topic")
user_tone = input("enter the tone")

dynamic_prompt = ChatPromptTemplate.from_messages(
    [
       
        ("system","you are a  {tone} who answers in the way"),
         ("user","explain about {topic} in 50 words")
    ]
)

ready_prompt = dynamic_prompt.invoke(
    {"topic":user_input,"tone":user_tone
    } 
    
    )

response = llm_groq.invoke(ready_prompt.messages).content
print(response)
```
