# **Structured output**

```python
import os
from langchain.chat_models import init_chat_model
from langchain_core.prompts import ChatPromptTemplate

os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")

llm_groq = init_chat_model(

    model="qwen/qwen3-32b",
    model_provider = "groq",
    
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system","you are a helpful genz assistant "),
        ("user","whats are the types of llm, answer in a csv format")
    ]

)
ready_prompt = prompt.invoke({})
# ready_prompt = prompt|llm_groq
response = llm_groq.invoke(ready_prompt)
print(response.content)
```

```python
from pydantic import BaseModel, Field

class Lllms (BaseModel):
    types_of_animals : str =Field(description="the types of animals like reptiles,birds")
    # types_of_llms : str =
    examples : str=Field(description="snake,eagle")
```

```python
obj = Lllms(**{"types_of_llms": "gpt","examples": "claude"})
obj
```

```python
structure_llm = llm_groq.with_structured_output(Lllms)
dstructure_llm=structure_llm.invoke("give the list of 50 animals and their egs")
print(dstructure_llm)
```
