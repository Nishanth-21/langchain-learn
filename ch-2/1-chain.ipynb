```python
# task 2


llm_groq = init_chat_model(
    model="qwen/qwen3-32b",
    model_provider="groq",
    temperature = 0.9
)
```

```python
from langchain_core.output_parsers import StrOutputParser

# llm_groq.invoke("what is a banana ? ")
# respponse.content

strr = StrOutputParser()

chain = prompt | llm_groq | strr

chain.invoke("what is the love")

```

```python
import re



prompt = ChatPromptTemplate.from_messages([
    ("system", "u are a helpful nutrionist "),
    ("user","based on my {age} and {weight} give me suggestions")
])


chain = prompt | llm_groq | strr

response = chain.invoke({
    "age":21,
    "weight":50
})


response = re.sub(r"<think>.*?</think>\s*", "", response, flags=re.DOTALL)
print(response)
```

```python
from langchain_core.runnables import RunnableSequence
import re


prompt = ChatPromptTemplate.from_messages([
    ("system", "u are a helpful nutrionist "),
    ("user","{question}")
])


chain_1 = RunnableSequence(
    prompt, llm_groq, strr
)

response = chain_1.invoke("what is meditation in simple terms ")
response = re.sub(r"<think>.*?</think>s/*","",response,flags=re.DOTALL)
print(response)
```
